```python
df.sort_values(by='democracy_score', ascending=False)['country'].head()
```

```python
df['region'].value_counts().count()
```

```python
Eastern_Europe = df['region'] == 'Eastern Europe'
df[Eastern_Europe]['country']
```

```python
population_2 = df['population'] == df['population'].nlargest(2).iloc[1]
df[population_2]['political_leader']
```

```python
leader_null = df['political_leader'].isna()
df[leader_null]['country'].count()
```

```python
def check_republic(txt):
  if 'Republic' in txt:
    return True
  else:
    return False

republic = df['country_long'].apply(check_republic)
df[republic]
```

```python
africa = df['continent'] == 'Africa'
africa_max = df[africa]['population'] == df[africa]['population'].max()
df[africa][africa_max]
```
