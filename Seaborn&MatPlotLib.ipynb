```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
y = x**2
plt.plot(x, y)
plt.title("Hello")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x,x)
plt.subplot(2,2,4)
plt.plot(y,y)
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
plt.savefig("plot1.png")
axis2 = fig.add_axes([1,0.3,0.5,0.5])
axis2.plot(y,x)
```

```python
import seaborn as sns
```

```python
df1 = sns.load_dataset('tips')
df1.head()
```

```python
df1['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df1['total_bill'], bins=20, kde = True)
plt.subplot(1,2,2)
sns.histplot(df1['tip'], bins=20, kde = True)
```

```python
sns.jointplot(x='total_bill', y='tip', data = df1, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data = df1, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data = df1, kind='reg')
```

```python
sns.pairplot(df1, hue='size', palette='rainbow')
```

```python
sns.rugplot(df1)
```

```python
sns.countplot(x = df1['sex'], hue = df1['smoker'])
```

```python
sns.barplot(x = df1['sex'], y = df1['total_bill'])
```

```python
sns.barplot(x = df1['sex'], y = df1['tip'], estimator=np.sum)
```

```python
sns.boxplot(x = 'tip', y = 'day', data = df1)
```

```python
sns.violinplot(x = 'day', y='total_bill', data = df1)
```

```python
sns.stripplot(x = 'day', y='total_bill', data=df1)
```

```python
sns.swarmplot(x = 'day', y='total_bill', data=df1)
```

```python
sns.violinplot(x = 'tip', y='day', data=df1)
sns.swarmplot(x = 'tip', y='day', data=df1)
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips_corr = tips[['total_bill','tip','size']]
tips_corr
```

```python
corr1 = tips_corr.corr()
corr1
```

```python
sns.heatmap(corr1, annot=True)
```

```python
sns.clustermap(corr1)
```

```python
pvt = flights.pivot_table(values='passengers', index='month', columns='year')
pvt
```

```python
sns.heatmap(pvt)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue='sex', markers=['o', 'v'])
```
