```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('anime.csv')
df.head(10)
```

```python
def extract_ep(txt):
  check = False
  ep = ""
  for i in txt:
    if i == ')':
      break
    if check == True:
      ep += i
    if i == '(':
      check = True

  return ep

df['Episodes'] = df['Title'].apply(extract_ep)
df['Episodes'] = df['Episodes'].str.replace(" eps", "")
df['Episodes'] = df['Episodes'].astype(int)
df.head()
```

```python
df.loc[3]['Title']
```

```python
def extract_time(txt):
  check = False
  count = 19
  time = ""
  for i in txt:
    if (check == True) and (count > 0):
      time += i
      count -= 1
    if i == ')':
      check = True
  return time

df["Time_Line"] = df['Title'].apply(extract_time)
df.head()
```

```python
import pandas as pd

def get_months(duration):
    start, end = duration.split(" - ")

    start = pd.to_datetime(start)
    end = pd.to_datetime(end)

    months = (end.year - start.year) * 12 + (end.month - start.month)

    return months

df['Months'] = df['Time_Line'].apply(get_months)
df.head()
```
